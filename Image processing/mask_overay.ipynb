## Segmentation: mask와 data 표시하기

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data = cv2.imread('data.png')[...,::-1]
```

```python
mask = cv2.imread('mask.png',0)
```

```python
fig,ax = plt.subplots(1,2)
ax[0].imshow(data)
ax[1].imshow(mask)
plt.show()
```

###  matplotlib

```python
#plot pred mask with red color
g= b = np.zeros_like(mask)
pred_mask= cv2.merge([mask,g,b])
```

```python
fig,ax = plt.subplots()
ax.imshow(data)
plt.imshow(pred_mask, alpha=0.1)
plt.show()
```

### opencv

```python
color =[255,0,0]
alpha=0.2
```

```python
colored_mask = np.expand_dims(mask, 2).repeat(3, axis=2)
masked = np.ma.MaskedArray(data, mask=colored_mask, fill_value=color)
image_overlay = masked.filled()

image_combined = cv2.addWeighted(data, 1 - alpha, image_overlay, alpha, 0)
```

```python
plt.imshow(image_combined)
```
